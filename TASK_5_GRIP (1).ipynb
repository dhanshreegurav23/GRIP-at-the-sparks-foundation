```
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
print('Libraries Imported')
df=pd.read_csv('/content/SampleSuperstore.csv')
df
```

```
#Removing duplicate values:
df=df.drop_duplicates()
```

```
df['Country'].value_counts()
```

```
df.drop('Country',axis=1,inplace=True)
df.drop('Postal Code',axis=1,inplace=True)
df
```

```
#correlation matrix:
plt.figure(figsize=(10,6))
correlation=df.corr()
sns.heatmap(correlation,annot=True,cmap='Blues')
```

```
plt.figure()
sns.pairplot(df)
```

```
plt.figure(figsize=(10,5))
sns.lineplot(x='Discount',y='Profit',data=df)
plt.legend('Profit')
plt.grid(alpha=0.3)
```

```
plt.figure(figsize=(10,5))
plt.plot(df['Profit'], color='Pink')
plt.grid(alpha=0.3)
```

```
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.countplot('Segment',data=df)
plt.title('DISTRIBUTION OF SEGMENTS')
plt.xlabel('SEGMENT',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.grid(alpha=0.3)
```

```
plt.figure(figsize=(10,5))
plt.subplot(1,2,2)
sns.countplot('Ship Mode',data=df)
plt.title('DISTRIBUTION OF SHIP MODE')
plt.xlabel('SHIP MODE',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.grid(alpha=0.3)
```

1. From the above "DISTRIBUTION OF SEGMENTS" plot it is clear that the "consumers" are the highest buyers of goods and "Home office" are the least buyers of goods
2. From the above "DISTRIBUTION OF SHIP MODE" plot, it is clear that the "standard class" mode is the highest mode for shipping goods and "same day" mode is the least mode for shipping goods

```
plt.figure(figsize=(30,10))
sns.countplot(df['State'])
plt.xticks(fontsize=16,rotation=90)
plt.ylabel('COUNTS',fontsize=20)
plt.xlabel('STATES',fontsize=20)
plt.title('DISTRIBUTION OF STATES',fontsize=25);
plt.grid(alpha=0.3)
```

From observing the above distribution plot, it is clear that the goods shipped to California is higher than any other states in the United States, in other words California has the highest rate of buyers comparing to other states and West Virginia has the least rate of buyers in the entire United States

```
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.countplot('Region',data=df)
plt.title('DISTRIBUTION Base on REGIONS')
plt.xlabel('REGION',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.grid(alpha=0.3)
```

```
plt.figure(figsize=(10,6))
plt.subplot(1,2,2)
sns.countplot('Region',data=df)
plt.title('DISTRIBUTION BASE ON CATEGORY')
plt.xlabel('REGION',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.grid(alpha=0.3)
```

From the above plots it is clear that the Western states has the highest rate of buyers while southern states has the lowest rate of buyers. From our previous observations we found out that the California has the highest rate of buyers - in this case California also a part of western states and this proves our statement.

```
plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
sns.countplot('Category',data=df);
plt.title('DISTRIBUTION OF CATEGORY')
plt.xlabel('CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
```

```
plt.figure(figsize=(10,6))
plt.subplot(2,1,2)
sns.countplot('Sub-Category',data=df);
plt.title('DISTRIBUTION OF SUB-CATEGORY')
plt.xlabel('SUB-CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
```

From the above plots it is clear that the office supplies has the highest rate of shippings in total. Here, Binders has the highest shipping count while copiers has the least shipping count under office supplies
